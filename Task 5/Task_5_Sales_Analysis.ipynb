```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import random

products = ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard']
regions = ['North', 'South', 'East', 'West', 'Central']

data = {
    'Product': [random.choice(products) for _ in range(30)],
    'Region': [random.choice(regions) for _ in range(30)],
    'Sales': [random.randint(1000, 5000) for _ in range(30)]
}


df = pd.DataFrame(data)

df.to_csv('sales.csv', index=False)
df.head(10)
```

```python
# Total sales by Product
product_sales = df.groupby('Product')['Sales'].sum()

# Total sales by Region
region_sales = df.groupby('Region')['Sales'].sum()

print("Product-wise Sales:\n", product_sales)
print("\nRegion-wise Sales:\n", region_sales)
```

```python
# Bar chart: Product-wise sales
product_sales.plot(kind='bar', color='lightgreen', title='Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

# Pie chart: Region-wise sales
region_sales.plot(kind='pie', autopct='%1.1f%%', title='Sales by Region')
plt.ylabel('')
plt.show()
```

### 🔍 Insights:
- Product C is leading in total sales.
- North region contributes the highest share to the sales.
